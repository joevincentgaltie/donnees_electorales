### Import

```python
import pandas as pd 
import numpy as np
```

```python
#fonctions utiles poiru présidentielles 

Base = ['Code du département', 'Libellé du département',
       'Code de la circonscription', 'Libellé de la circonscription',
       'Code de la commune', 'Libellé de la commune', 'Code du b.vote',
       'Inscrits', 'Abstentions', '% Abs/Ins', 'Votants', '% Vot/Ins',
       'Blancs', '% Blancs/Ins', '% Blancs/Vot', 'Nuls', '% Nuls/Ins',
       '% Nuls/Vot', 'Exprimés', '% Exp/Ins', '% Exp/Vot']
Candidat = [ 'N°Panneau', 'Sexe',
       'Nom', 'Prénom', 'Voix', '% Voix/Ins', '% Voix/Exp']

#function to fill the columns index,

def fill_columns_index (shape) : 
    columns = Base.copy() 
    i = 0
    while len(columns) < shape : 
        #adding the candidat columns plus identifier 
        i += 1
        columns.extend([ ele + str(i) for ele in Candidat])
    return columns



#extracting Base data, and candidates data one candidate by one candidate  and merge it iteratively to a single dataframe
def extract_data(df, shape) :
    columns = fill_columns_index(shape)
    df.columns = columns
    df_base = df[Base]
  #iterating over 7 columns of candidates data starting from 19th column
    for i in range(21, shape, 7):
        df_candidat = df[columns[i:i+7]]
        df_candidat.columns = Candidat
        df_candidat = df_candidat.fillna(method='ffill')
        df_candidat = df_candidat.fillna(method='bfill')
        df_candidat = df_candidat.fillna(0)
        df_candidat = df_candidat.reset_index(drop=True)
        df_candidat = df_candidat.drop(['N°Panneau'], axis=1)
        df_candidat = df_candidat.drop(['Sexe'], axis=1)
        if i == 21:
            df_new= pd.concat([df_base, df_candidat], axis=1)
        else:
            df_new = pd.concat([df_new, pd.concat([df_base, df_candidat], axis=1)], axis=0)
    
    return df_new
```
